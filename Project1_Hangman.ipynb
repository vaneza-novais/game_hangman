```python
#Project 1 - Dev. game Handgame
import random as rd
import pandas as pd
import numpy as np

list_word = ["table", "chair", "stove", "sofa", "tv", "fridge", "closet", "bed", "pillow", "stair"]
list_blank = []

#variables
word = rd.randint(0, len(list_word))
word = list_word[word]
list_blank_res = []

word = "pillow"

attempt_max = len(word)*2

#show
result = "_ "*len(word)

print(f"Get the word right: \n {result} \n\nYou have {attempt_max} chance")

list_blank = ["_" for i in word]

#to check if there duplicates
def check_duplicates (word):
    if len(word) > len(set(word)):
        repetead = {i for i in word if word.count(i) > 1}
    else:
        repetead = []
    return repetead
    
repetead = check_duplicates(word)

for i in range (1,attempt_max + 1):
    input_user = input("What letter? \n").lower()
    
    #to check if is string
    if input_user.isalpha():
        #to check if letter is in the word
        if input_user in word:
            for position, element in enumerate(word):
                if element == input_user:
                    list_blank.pop(position) #delete _
                    list_blank.insert(position, input_user) #add letter          

            list_blank_res = "".join(list_blank)

            if list_blank_res == word:
                print(f"Very good, you won. \nThe word is: {word}")

            else:
                print(f"Very good, keep it up \n{list_blank}")
                                                   
        if list_blank_res == word:
             break

        elif (input_user not in word) and i < attempt_max:
            print("Error, try again")

        elif (input_user not in word) and i == attempt_max:
            print (f"You lost. \nThe word is: {word}")
        
        else:
            pass
            
    else:
        print("Enter a string. \nOne less chance.")  

```
